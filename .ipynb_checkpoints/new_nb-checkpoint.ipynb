### Load and preprocess the data

```python
# Load data
data = pd.read_excel("data/Realized Schedule 20210101-20220208.xlsx")

#############################
########Preprocessing########
#############################

# Remove rows with missing values
df = data.dropna()

# Remove row with flight type G
df = df[df["FlightType"] != "G"]


#Convert schedule time to different columns
# Include month, year, day and weekday in the dataset
df = pd.concat([df, pd.to_datetime(df.ScheduleTime).dt.year.rename("Year")], axis=1)
df = pd.concat([df, pd.to_datetime(df.ScheduleTime).dt.month.rename("Month")], axis=1)
df = pd.concat([df, pd.to_datetime(df.ScheduleTime).dt.week.rename("Week")], axis=1) # Remove isocalendar if error
df = pd.concat([df, pd.to_datetime(df.ScheduleTime).dt.day.rename("Day")], axis=1)
df = pd.concat([df, pd.to_datetime(df.ScheduleTime).dt.weekday.rename("Weekday")], axis=1)
df = pd.concat([df, pd.to_datetime(df.ScheduleTime).dt.hour.rename("Hour")], axis=1)

# Include dummy variables for categorical features
cols = ["Airline", "Destination", "FlightType", "Sector"]
col_prefix = ["AIR", "DEST", "FLT", "SECT"]
weekdays = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
cols_to_drop = ["FlightNumber", "AircraftType", "ScheduleTime", "Weekday"]

for i, col in enumerate(cols):
    df = pd.concat([df, pd.get_dummies(df[col], prefix=col_prefix[i])], axis=1)

weekday_df = pd.get_dummies(df.Weekday)
weekday_df.columns = weekdays

df = pd.concat([df, weekday_df], axis=1)

df.drop(cols, axis=1, inplace=True)
df.drop(cols_to_drop, axis=1, inplace=True)

print(df.columns)
print(df.shape)

```

```python
# Extract features and targets
X = df.drop(["LoadFactor"], axis=1)
y = df.LoadFactor

# Perform train/val split
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.25, random_state=1405)
```

### Model selection part
In this part we try to fit different models to our data : 
1. Forward selection
2. Backward selection
3. Decision tree
4. Random forest
5. K-Nearest Neighbours


#### Forward selection

```python
# Extract features using forward selection
forward_selection_features = forward_selection(X_train, y_train, num_of_features=10)
print(X_train.columns[forward_selection_features])
```

```python
# Perform linear regression with the forward selected features
X_forward_sel = X_train.iloc[:, forward_selection_features]
X_val_sel = X_val.iloc[:, forward_selection_features]

model = LinearRegression()
model.fit(X_forward_sel, y_train)

y_hat = model.predict(X_val_sel)
print(y_hat[:10])
print(y_val[:10].to_numpy())

RMSE = np.sqrt(np.mean((y_hat - y_val)**2))
print(f'RMSE for forward selected model is: {RMSE}')
print(f'Lengths:\nyhat: {len(y_hat)}, yval: {len(y_val)}')

# Visualize the results
plt.plot(y_val, 'bo')
plt.plot(y_hat, 'go')

print("\nMSE of LR with forward selection :", mean_squared_error(y_val, y_hat))
```

```python
num_features = [i*10 for i in range(1,6)]
n = len(num_features)
train_MSE_fs = [None for _ in range(n)]
val_MSE_fs = [None for _ in range(n)]
sel_features = [None for _ in range(n)]

model_fs = LinearRegression()

for i,num in enumerate(num_features):
    fs_features = forward_selection(X_train, y_train, num_of_features=num)
    X_train_fs = X_train.iloc[:, fs_features]
    X_val_fs = X_val.iloc[:, fs_features]
    
    model_fs.fit(X_train_fs, y_train)
    y_hat_fs = model_fs.predict(X_val_fs)
    
    train_MSE_fs[i] = mean_squared_error(model_fs.predict(X_train_fs), y_train)
    val_MSE_fs[i] = mean_squared_error(y_val, y_hat_fs)
    sel_features[i] = fs_features
    
    print(X_train.columns[fs_features])
    print(f"Finished step {i+1}/{len(num_features)}")
```

```python
print(num_features)
```

```python
fig, ax = plt.subplots(1, 1)
plt.plot(num_features, val_MSE_fs, 'o-', label='Validation')
plt.plot(num_features, train_MSE_fs, 'o-', label='Training')
plt.legend()
ax.set_xlabel("Number of selected features")
ax.set_ylabel("Training and validation MSE")
plt.show()
```

#### Backward selection

```python
# Extract features using backward selection
model = LinearRegression()
f_selector = RFE(model, n_features_to_select=10, step=1)
f_selector.fit(X_train, y_train)
print(X.columns[f_selector.support_])
```

```python
backward_selected_features = X_train.columns[f_selector.support_]
print(backward_selected_features)

X_backward_sel = X_train.loc[:, backward_selected_features]
X_val_sel = X_val.loc[:, backward_selected_features]

model = LinearRegression()
model.fit(X_backward_sel, y_train)

y_hat_BS = model.predict(X_val_sel)
print(y_hat_BS[:10])
print(y_val[:10].to_numpy())

RMSE = np.sqrt(np.mean((y_hat_BS - y_val)**2))

print(f'RMSE for forward selected model is: {RMSE}')
print(f'Lengths:\nyhat: {len(y_hat_BS)}, yval: {len(y_val)}')

print("\nMSE of LR with backward selection :", mean_squared_error(y_val.to_numpy(), y_hat_BS))
```

#### Decision tree

```python
dtree = DecisionTreeRegressor()
dtree.fit(X_train, y_train)

y_hat_DT = dtree.predict(X_val)

print("MSE of decision tree :", mean_squared_error(y_val.to_numpy(), y_hat_DT))
print("Accuracy of decision tree :", accuracy(y_hat_DT, y_val.to_numpy()))
```

#### Calculate the deviation 

```python
accuracy(y_hat_DT, y_val)
```

#### Random forest

```python
rd_forest = RandomForestRegressor(bootstrap=True, oob_score=True, random_state=0)

rd_forest.fit(X_train, y_train)
y_hat_rdf = rd_forest.predict(X_val)

print("MSE of random forest :", mean_squared_error(y_val.to_numpy(), y_hat_rdf))
print("Accuracy of random forest :", accuracy(y_hat_rdf, y_val.to_numpy()))
```

#### KNN

```python
K = 20
KNN = KNeighborsRegressor(n_neighbors = K, weights = 'uniform', metric = 'euclidean')
KNN.fit(X_train, y_train)

pred_knn = KNN.predict(X_val)

print("MSE of KNN :", mean_squared_error(y_val.to_numpy(), pred_knn))
print("Accuracy of KNN :", accuracy(pred_knn, y_val.to_numpy()))
```

```python
Ks = [1, 2, 5, 10, 20, 30, 50, 70, 100]
val_errors = [None for _ in range(len(Ks))]
train_errors = [None for _ in range(len(Ks))]

val_acc = [None for _ in range(len(Ks))]
train_acc = [None for _ in range(len(Ks))]

for i,k in enumerate(Ks):
    KNN = KNeighborsRegressor(n_neighbors = k, weights = 'uniform', metric = 'euclidean')
    KNN.fit(X_train, y_train)
    pred_knn = KNN.predict(X_val)

    val_errors[i] = mean_squared_error(y_val.to_numpy(), pred_knn)
    train_errors[i] = mean_squared_error(y_train.to_numpy(), KNN.predict(X_train))
    
    val_acc[i] = accuracy(pred_knn, y_val.to_numpy())
    train_acc[i] = accuracy(KNN.predict(X_train), y_train.to_numpy())
```

```python
fig, axs = plt.subplots(1, 2, figsize=(15,6))

axs[0].plot(Ks, val_errors, 'o-', label='Validation')
axs[0].plot(Ks, train_errors, 'o-', label='Training')
axs[0].legend()
axs[0].set_xlabel("Number of neighbors")
axs[0].set_ylabel("Training and validation MSE")

axs[1].plot(Ks, val_acc, 'o-', label='Validation')
axs[1].plot(Ks, train_acc, 'o-', label='Training')
axs[1].legend()
axs[1].set_xlabel("Number of neighbors")
axs[1].set_ylabel("Training and validation accuracies")


plt.show()
```

A more precise selection of k value : 

```python
Ks = [k for k in range(2,16)]
val_MSE = [None for _ in range(len(Ks))]
val_acc = [None for _ in range(len(Ks))]
train_errors = [None for _ in range(len(Ks))]

for i,k in enumerate(Ks):
    KNN = KNeighborsRegressor(n_neighbors = k, weights = 'uniform', metric = 'euclidean')
    KNN.fit(X_train, y_train)
    pred_knn = KNN.predict(X_val)

    val_MSE[i] = mean_squared_error(y_val.to_numpy(), pred_knn)
    val_acc[i] = accuracy(pred_knn, y_val.to_numpy())
    train_errors[i] = mean_squared_error(y_train.to_numpy(), KNN.predict(X_train))
```

```python
print("Value of K with the smallest error :", Ks[np.argmin(val_MSE)])
print("Value of K with the highest accuracy :", Ks[np.argmax(val_acc)])

fig, axs = plt.subplots(1, 2, figsize=(15, 6))
axs[0].plot(Ks, val_MSE, 'o-', label='Validation')
axs[0].legend()
axs[0].set_xlabel("Number of neighbors")
axs[0].set_ylabel("Validation MSE")

axs[1].plot(Ks, val_acc, 'o-', label='Validation')
axs[1].legend()
axs[1].set_xlabel("Number of neighbors")
axs[1].set_ylabel("Validation accuracy")

plt.show()
```

The selected value of k depends on the validation set. It varies from 7 to 12 usually.

### Predictions based only on the same month, year before

Try KNN to predict based on the data of the same month, a year ago

```python
# Prepare train and test

january_train = df[(df.Month == 1) & (df.Year == 2021)]
january_val = df[(df.Month == 1) & (df.Year == 2022)]
febuary_train = df[(df.Month == 2) & (df.Year == 2021)]
febuary_val = df[(df.Month == 2) & (df.Year == 2022)]

X_train_j = january_train.drop(["LoadFactor"], axis=1)
y_train_j = january_train.LoadFactor
X_val_j = january_val.drop(["LoadFactor"], axis=1)
y_val_j = january_val.LoadFactor

X_train_f = febuary_train.drop(["LoadFactor"], axis=1)
y_train_f = febuary_train.LoadFactor
X_val_f = febuary_val.drop(["LoadFactor"], axis=1)
y_val_f = febuary_val.LoadFactor
```

```python
K = 8
KNN_j = KNeighborsRegressor(n_neighbors = K, weights = 'uniform', metric = 'euclidean')
KNN_j.fit(X_train_j, y_train_j)

pred_knn_j = KNN_j.predict(X_val_j)

print("MSE of KNN january :", mean_squared_error(y_val_j.to_numpy(), pred_knn_j))

KNN_f = KNeighborsRegressor(n_neighbors = K, weights = 'uniform', metric = 'euclidean')
KNN_f.fit(X_train_f, y_train_f)

pred_knn_f = KNN_f.predict(X_val_f)

print("MSE of KNN febuary :", mean_squared_error(y_val_f.to_numpy(), pred_knn_f))
```

```python
dtree = DecisionTreeRegressor()

dtree.fit(X_train_j, y_train_j)
y_hat_j = dtree.predict(X_val_j)
print("MSE of decision tree january:", mean_squared_error(y_val_j.to_numpy(), y_hat_j))

dtree.fit(X_train_f, y_train_f)
y_hat_f = dtree.predict(X_val_f)
print("MSE of decision tree febuary:", mean_squared_error(y_val_f.to_numpy(), y_hat_f))
```

```python
rd_forest = RandomForestRegressor(bootstrap=True, oob_score=True, random_state=0)

rd_forest.fit(X_train_j, y_train_j)
y_hat_rdf_j = rd_forest.predict(X_val_j)
print("MSE of random forest january :", mean_squared_error(y_val_j.to_numpy(), y_hat_rdf_j))

rd_forest.fit(X_train_f, y_train_f)
y_hat_rdf_f = rd_forest.predict(X_val_f)
print("MSE of random forest febuary :", mean_squared_error(y_val_f.to_numpy(), y_hat_rdf_f))
```
